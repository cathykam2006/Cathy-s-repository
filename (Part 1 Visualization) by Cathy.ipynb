```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('COVID-19_Vaccination_Age_and_Sex_Trends_in_the_United_States__National_and_Jurisdictional.csv')
```

# The overall distribution of population of completing entire vaccination series

```python
df.drop(df[df.Location == 'US'].index, inplace=True)
```

```python
fig, ax = plt.subplots(figsize=(20,10))
sns.barplot(data = df.sort_values(['Series_Complete_Yes'], ascending = False), y = 'Series_Complete_Yes', 
            x = 'Location');
plt.ylabel('Population Per 10 million')
plt.xlabel('Locations')
plt.title('The Overall Distribution of Entire Vaccination Series completed');
```

# The overall distribution of Completing Adminstered Dose 1

```python
fig, ax = plt.subplots(figsize=(20,10))
sns.barplot(data = df.sort_values(['Administered_Dose1'], ascending = False), y = 'Administered_Dose1', 
            x = 'Location');
plt.ylabel('Population Per Million')
plt.xlabel('Locations')
plt.title('The Overall Distribution of Completing Adminstered Dose 1');
```

# The overall distribution of Booster Doses across different locations

```python
fig, ax = plt.subplots(figsize=(20,10))
sns.barplot(data = df.sort_values(['Booster_Doses'], ascending = False), y = 'Booster_Doses', 
            x = 'Location')
plt.ylabel('Population Per Million')
plt.xlabel('Locations')
plt.title('The Overall Distribution of Completing 1st Booster Doses');
```

```python
fig, ax = plt.subplots(figsize=(20,10))
sns.barplot(data = df.sort_values(['Second_Booster'], ascending = False), y = 'Second_Booster', 
            x = 'Location')
plt.ylabel('Population Per Million')
plt.xlabel('Locations')
plt.title('The Overall Distribution of Completing 2nd Booster Doses');
```

```python
fig, ax = plt.subplots(figsize=(40,30))
sns.barplot(data = df.sort_values(['Demographic_Category'], ascending = True), x = 'Series_Complete_Yes', 
            y = 'Demographic_Category')
plt.ylabel('Age Demographic')
plt.xlabel('Population per million')
plt.xticks(fontsize = 20)
plt.title('The Overall Demographic of Completing the Entire Vaccination Series');
```

```python
sns.barplot(x= df.Demographic_Category.str.contains('Female'), y= df.Series_Complete_Yes, ax=ax, label="Entire Vaccination series completed")
```

```python
sns.set(style="whitegrid")
  
ax = sns.scatterplot(x='Series_Complete_Yes', y='census', data=df)

ax.set(xlabel ='Population Completed Entire Vaccination (per 10 million)', ylabel ='Census Population (per 10 million)')

plt.title('The Relationship between Completion of Entire Vaccination and the Census', fontsize=15)

plt.show()
```

```python
sns.set(style="whitegrid")
  
ax = sns.scatterplot(x='Administered_Dose1', y='Booster_Doses', data=df, hue = 'Booster_Doses')

ax.set(xlabel ='Population Completed Dose 1 per 10 million', ylabel ='Population received Booster Doses per 10 million')

plt.title('The Relationship between Completion of Dose 1 and Booster Doses', fontsize=15)

plt.show()
```

```python
fig, ax = plt.subplots(figsize=(10,6))

sns.set(style="whitegrid")
  
sns.scatterplot(x= df['Date'], y=df['Second_Booster'], data=df, hue = df['Second_Booster'])

#ax.set(xlabel =df['Date'], ylabel ='Second Booster (per 10 million)')

ax.set_xticks(np.arange(0, df.shape[0] , 5))

plt.title('The Completion Second Booster From 12/13/2020 to 2/07/2022', fontsize=15)

plt.show()
```

```python
sns.set(style="whitegrid")
  
ax = sns.scatterplot(x='Date', y='Administered_Dose1', data=df, hue = 'Administered_Dose1')

ax.set(xlabel ='Date', ylabel ='Administered Dose 1 (per 10 million)')

plt.title('The Completion of Administered Dose 1 From 12/13/2020 to 2/07/2022', fontsize=15)

plt.show()
```

```python
sns.set(style="whitegrid")
  
ax = sns.scatterplot(x='Date', y='Booster_Doses', data=df, hue = 'Booster_Doses')

ax.set(xlabel ='Date', ylabel ='Booster Doses (per 10 million)')

plt.title('The Completion of Booster Doses From 12/13/2020 to 2/07/2022', fontsize=15)

plt.show()
```
